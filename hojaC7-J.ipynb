```python
import pandas as pd
import numpy as np
import seaborn as sns
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import requests
import matplotlib.pyplot as ptl
```

#### Primero scrapeando una hoja del archivo excel "Genes identificados con vecinos"

```python
gen_abrev=[]
gen_abrev2=[]
DF = pd.read_excel('Genes identificados con vecinos.xlsx',sheet_name='C7-J')
genabrev=DF["Gen Abrev"]
print(genabrev)

for i in genabrev:
    i=i.replace(" ","+")
    gen_abrev2.append(i)

for i in gen_abrev2:
    gen_abrev.append("https://www.uniprot.org/uniprot/?query="+i+"+bos+taurus&sort=score")
print(gen_abrev)
```

```python
lista=[]
for i in gen_abrev:
        urlv=i
        uclient=urlopen(urlv)
        pagehtml=uclient.read()
        uclient.close()
        pagesoup=soup(pagehtml, "html.parser")
    
        if pagesoup.find("div",{"id":"noResultsMessage"}):
            lista.append("NA") #no results found
        else:    
            lista.append(urlv)
print(lista)    
```

```python
ENTRY=[]
ENTRYNAME=[]

for i in lista:
    url=i   
    
    if url=="NA":
        ENTRY.append(url) #no results found
        ENTRYNAME.append(url)
    else:    
        uclient=urlopen(url)
        pagehtml=uclient.read()
        uclient.close()
        pagesoup=soup(pagehtml, "html.parser")
        page=pagesoup.find(class_="entry selected-row")
        
        entry=page.find("td",{"class":"entryID"})
       
        ENTRY.append(entry.find("a").text)
        ENTRYNAME.append(page.findAll("td")[2].text)
```

```python
link=[]
for i in ENTRY:
    link.append("https://www.uniprot.org/uniprot/"+i)

links=[]        
for i in link:
        urln=i
    
        if urln=="https://www.uniprot.org/uniprot/NA":
            links.append("NA") #no results found
        else:    
            links.append(urln)
print(links)  
```

```python
PROTEIN=[]
GENE=[]
ORGANISM=[]
STATUS=[]
GOBIOLOGICAL=[]
BIO=[]

for i in links:
    url=i   
    
    if url=="NA":
        PROTEIN.append(url) #no results found
        GENE.append(url)
        ORGANISM.append(url)
        STATUS.append(url)
        GOBIOLOGICAL.append(url)
    else:    
        uclient=urlopen(url)
        pagehtml=uclient.read()
        uclient.close()
        pagesoup=soup(pagehtml, "html.parser")
        page=pagesoup.find("div",{"id":"entry-overview"})
        page2=pagesoup.find("ul",attrs={"class":"noNumbering biological_process"})
        
        prot=page.find("div",{"id":"content-protein"})
        gen=page.find("div",attrs={"id":"content-gene"})
        organ=page.find("div",{"id":"content-organism"})
        stat=page.find("div",{"id":"content-status"})
        
        if page2 == None:
            GOBIOLOGICAL.append("NA") #no results found        
        else:
            page3=page2.findAll("a",{"onclick":"window.ga('UniProt-Entry-View', 'click', 'Display-GO-Term');"})
            page4=[]
            for i in page3:
                page4.append(i.text)
            BIO.append(".".join(page4))
            GOBIOLOGICAL=BIO
               
        PROTEIN.append(prot.find("h1").text)
        GENE.append(gen.text)
        ORGANISM.append(organ.find("em").text)
        STATUS.append(stat.find("span").text)
```

```python
#Data frame:
hojaC7J=pd.DataFrame({"Gen Abrev":genabrev,"Entry":ENTRY,"Entry name":ENTRYNAME,"Protein":PROTEIN,"Gene":GENE,"Organism":ORGANISM,"Status":STATUS,"GoBiological":GOBIOLOGICAL},index=list(range(0,len(gen_abrev)))) 
#Convirtiendo a csv:
hojaC7J.to_csv("hojaC7-J.csv")
```
